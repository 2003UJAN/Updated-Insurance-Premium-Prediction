```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("/content/india_insurance_dataset.csv")
```

```
df.head()
```

```
# Distributions
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
sns.histplot(df['annual_premium'], bins=50, ax=ax[0])
ax[0].set_title("Premium Distribution")
sns.boxplot(x='city_tier', y='annual_premium', data=df, ax=ax[1])
ax[1].set_title("Premium by City Tier")
sns.scatterplot(x='risk_score', y='annual_premium', data=df.sample(5000), ax=ax[2])
ax[2].set_title("Premium vs Risk Score")
plt.tight_layout()
```

```
# Correlation
plt.figure(figsize=(10, 8))
num_cols = ['age','bmi','smoker','chronic_count','city_tier',
            'sum_insured','risk_score','premium_multiplier','annual_premium']
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
